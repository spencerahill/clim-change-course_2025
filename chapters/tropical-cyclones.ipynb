## Key concepts

### What is a {term}`tropical cyclone` (TC)?

Let's break down the two words:

**tropical**: originating in Earthâ€™s Tropics, i.e. the low latitudes

But not too low!  There's a big gap surrounding the equator in TC tracks that we'll see below soon

**cyclone**: a spinning storm, counterclockwise in the northern hemisphere, clockwise in the southern hemisphere

So together: a spinning storm originating in the Tropics.

In fact, there are many storm systems in the Tropics that are spinning to some degree.  But not all of these are tropical cyclones.

A tropical cyclone is a specific kind of spinning tropical storm with several key features.

### What is a {term}`hurricane`?  What is a {term}`typhoon`?  

The same thing!  Synonyms for tropical cyclone.

Hurricane: simply the name used in the U.S. and nearby regions for tropical cyclones (TCs)

Typhoon: simply the name used for tropical cyclones (TC) in the northwest Pacific


## Tropical cyclones as heat engines
### Background: {term}`Carnot cycle`
The **Carnot cycle** is a 4-stage cycle through which an engine can peform work.

It provides an upper bound on the *efficiency* of that work: in performing that work, what's the minimum possible heat lost?

We'll see that one can understand tropical cyclones as a kind of heat engine, and thus use the Carnot cycle to think about them.

#### Setup
The engine consists of a **mechanical piston**: imagine a cylinder that is open on one end. 

A disk that fits just inside the cylinder is placed inside it, creating a sealed chamber of air inside the cylinder.  

While the rest of the cylinder walls are perfectly rigid, meaning they don't move, the disk is able to move in and out.

Moving in, it reduces the volume of the chamber, which compresses the air inside.

Moving out, it expands the volume of the chamber, which expands the air inside.

The piston is in contact on one side with a **hot reservoir** and in contact on another side with a **cold reservoir**.

#### The four stages
The four stages are:

1. **isothermal expansion**: Heat is drawn from the hot reservoir into the chamber behind the piston.  

   However, the gas inside the chamber *stays at constant temperature*: that's the *isothermal* part.

   Instead of warming up, the heating causes the gas to *expand*, pushing the disk out and increasing the cylinder's *volume*.

   As such, the gas has done {term}`work` on the disk: its pressure pushes on the disk over the distance that the disk travels.
2. **adiabatic expansion**: The chamber expands further, but now *without any heat added or taken way*.

   The lack of heat exchange is what makes it {term}`adiabatic`.
   
   In the *isothermal* Stage 1, the expansion was driven by the heat being pumped in from the hot reservoir, which is how the temperature could stay unchanged.

   Now, this flow of heat has been shut off.  With no added heat but the chamber still expanding, the gas must *cool*.
3. **isothermal compression**: Heat is pumped *out* of the chamber into the cold reservoir.

   As in Stage 1, it is *isothermal*, meaning that the gas in the chamber stays at a fixed temperature.

   If the gas is at a fixed temperature but it is losing heat, it must compress.

   This causes the disk to move in, *compressing* the chamber in the process.
4. **adiabatic expansion**: The chamber compresses further, but the heat exchange with the cold reservoir has been shut off.
   
   With no heat exchange, if the volume decreases, then the temperature must *increase*.

   The resulting compression and warming bring the chamber back to its original pressure, volume, and temperature it had at the start of Stage 1.  And then the cycle repeats.

### The Carnot cycle applied to tropical cyclones
To do this, we ignore all of the motion *around* the center of the tropical cyclone and focus purely on the *radial* and *vertical* motions.

The *radial* motion is air moving inward toward the center or outward away from the center.

And, as you surely know by now, vertical motion is air moving up or down.

We do this because tropical cyclones are *mostly* circular: standing at the cyclone center, point yourself in any direction.  The variations moving along that direction are much larger than if you picked some distance from the center and moved in a circle around the center.

(The formal term for this is that they are approximately *axisymmetric*.)

Viewed in the resulting *radial-vertical* plane, the circulation within the tropical cyclone looks like one big closed loop with four distinct segments.

Start with air that's near the surface and radially far out, near the tropical cyclone's outer edge:

1. The air moves back radially inward, until it reaches the eyewall.  
2. At the eyewall, the air stops moving radially inward and begins rising vertically upward.
3. At the storm top, it stops moving upward but begins moving radially outward.
4. At the outer edge of the storm, it stops moving outward and instead starts sinking.  When it reaches near the surface, it is back where the first segment started, and the loop repeats.

Each of these four segments corresponds to one of the four stages of the Carnot cycle:

1. **isothermal expansion**: the air moving just above the warm tropical ocean surface is heated
2. **adiabatic expansion**: the air rises in moist convection, expanding in the process.  At least within the core of the updrafts, this is approximately adiabatic
3. **isothermal compression**: the air moves poleward near the tropopause.  Along the way, it is radiating away lots of energy.
4. **adiabatic compression**: the air descends, getting compressed in the process as it goes from lower to higher pressure.
